```python
import pandas as pd

# Chargement des bases
train = pd.read_csv("/content/Base_Entraînement - Feuille 1.csv")
test = pd.read_csv("/content/Base_Test - Feuille 1 (1).csv")

# Vérification des colonnes
assert list(train.columns) == list(test.columns), "Colonnes différentes"
print("Colonnes identiques")
```

```python
labels_train = set(train["Labels"].unique())
labels_test = set(test["Labels"].unique())

print("Labels train :", labels_train)
print("Labels test  :", labels_test)

assert labels_test.issubset(labels_train), "Une catégorie du test n’existe pas dans le train"
print("Toutes les catégories du test sont présentes dans le train")
```

```python
intersection = set(train["Texte"]).intersection(set(test["Texte"]))

assert len(intersection) == 0, "Fuite de données entre train et test"
print("Aucun texte dupliqué entre train et test")
```

```python
print("Répartition TRAIN :")
print(train["Labels"].value_counts())

print("\nRépartition TEST :")
print(test["Labels"].value_counts())
```
